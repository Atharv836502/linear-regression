# Linear regression

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\Admin\\OneDrive\\Desktop\\linear regression\\homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.Area,df.Price,color='Green',marker='*')
plt.show()
```

```python
new_df = df.drop('Price',axis='columns')
new_df
```

```python
price = df.Price
price
```

```python
#create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,price)
```

1) predict price of a home with area = 5000 sqr ft

```python
reg.predict([[5000]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
